# Task 1

Use the multi-channel image imgThresholding2.png

```python
p1 = io.imread('basics/imgThresholding2.png')

plt.title('Task 1 Image')
plt.imshow(p1)
plt.axis('off')
plt.show()
```

## Task 1a

Split it into individual channels (channel 0 - red, channel 1 - green)

```python
def pltrg(r, g):
    fig, ax = plt.subplots(1, 2)

    ax[0].set_title('Red Channel')
    ax[0].imshow(r, cmap=plt.cm.gray)

    ax[1].set_title('Green Channel')
    ax[1].imshow(g, cmap=plt.cm.gray)

    for a in ax.ravel():
        a.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
p1r, p1g = p1[:,:,0], p1[:,:,1]
pltrg(p1r, p1g)
```

## Task 1b

Obtain histograms of intensity values in these images.

```python
fig, ax = plt.subplots(1, 2)

ax[0].set_title('Red Histogram')
ax[0].hist(p1r)

ax[1].set_title('Green Histogram')
ax[1].hist(p1g)

#plt.tight_layout()
plt.show()
```

## Task 1c

Perform foreground-background detection by thresholding on the individual

```python
rt = filters.threshold_adaptive(p1r, 71)
#p1rb = p1r < rt

gt = filters.threshold_adaptive(p1g, 71)
#p1gb = p1g < gt

#pltrg(p1rb, p1gb)
pltrg(rt, gt)
```

```python
p1rb = filters.median(rt, selem=np.ones((5,5)))
p1gb = filters.median(gt, selem=np.ones((5,5)))

p1rb = morphology.binary_closing(rt)
p1gb = morphology.binary_closing(gt)

pltrg(p1rb, p1gb)
```

# Coin Demo

From scikit-image's website

```python
image = data.coins() # or any NumPy array!
io.imshow(image)
io.show()
```

```python
edges = filters.median(image, selem=np.ones((3, 3)))
io.imshow(edges)
io.show()
```

```python
threshold = filters.threshold_otsu(image)
binary = image > threshold + 22
io.imshow(binary)
io.show()
```

```python
closed = morphology.binary_closing(binary)
io.imshow(closed)
io.show()
```

```python
edges = filters.sobel(image)
io.imshow(edges)
io.show()
```
